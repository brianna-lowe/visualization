HOW MATPLOTLIB WORKS
# A figure is like a contaier that holds a set of AXES
# the axes is our actual plot or graph
# a figure cam hold multiple axes ( like subplots)
# every visual element of our plots - colour, legends, axis titles and scales, text is called an ARTISTS and belongs to an axes ( not to a figure)

```python
# activity:customizing our plots post-activity
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(9,3)) 
plt.xticks(rotation = 45, ha = 'right', fontsize = 12, font = "serif" )

ax1.scatter(x1, y1, marker = 'o', color = 'red')
ax2. bar(x2,y2, color = 'green')

ax1.set_title('Important Scatter Plot', fontdict=font1)
ax1.set_ylabel('# of Circles', fontdict=font2)
ax1.set_xlabel('Over Years', fontdict=font2)

ax2.set_title('Important Bar Graph', fontdict=font1)
ax2.set_xlabel('Strongest One Piece Character', fontdict=font3)
ax2.set_ylabel('# of KOs', fontdict=font3)
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3', ],
                                    ['ax2', 'ax3']],
                                    figsize =(7,4))
someaxes['ax1'].scatter(x1, y1)
someaxes['ax2'].bar(x2,y2)
someaxes['ax3'].plot(x1,y1)
```

```python
#adjusting subplot layouts
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3', ],
                                    ['ax2', 'ax3']],
                                    figsize =(7,4),
                                    layout = "constrained") #fixing the layout
someaxes['ax1'].scatter(x1, y1)
someaxes['ax2'].bar(x2,y2)
someaxes['ax3'].plot(x1,y1)


someaxes['ax1'].set_xlabel('A Big Label', fontsize=18)
someaxes['ax2'].set_xlabel('Another Label', fontsize=18)
someaxes['ax3'].set_xlabel('Label 2:2 Fast 2 Furious', fontsize=18)
```

```python
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"]) #catergorical data
y1 = np.array([110,180,240, 99, 220]) #numerical values as an array
y2 = np.array([170,100,90, 120, 50]) #numerical values as an array

fig, ax = plt.subplots(figsize=(7,3))
ax.bar(x,y1, color ='indigo')
ax.plot(x,y2, color ='red')

```

```python
#adding error information 
y2_sd = np.std(y2) # here it is #adding error information 

fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x,y2, color ='red')

ax.errorbar(x, y2, yerr= y2_sd, fmt= "none" ) #format equals none
```

```python
#customizing errorbar appearence
y2_sd = np.std(y2) # here it is #adding error information 

fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x,y2, color ='red')

ax.errorbar(x, y2, yerr= y2_sd, fmt= "none", ecolor='indigo', elinewidth=4, capsize=6, capthick=4 ) #format equals none
```
