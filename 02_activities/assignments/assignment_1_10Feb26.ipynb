```python
# generate fake data
np.random.seed(613) # ciara thinks this is a lucky number
X = np.arange(50) # number on the x axis - a range of 50 values
y = np.random.randint(0,100,50) # random int between 0-100 and we want 50 of them (total # of blue dots ARE 50!)

# plot our fake data
fig, ax = plt.subplots(figsize=(5,3))  # figure and put axes obj, size of fig be 5,3 inches - can be changed with code when exporting 
ax.hist(y) # how it looks and what data to plot #histogram is counting of the occurance of y value
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(X,y)
ax.set_title('Total Growth Over Time')
ax.set_ylabel('Total Growth')
ax.set_xlabel('Years Since Start')
fig.tight_layout() # keep everything fitting into container - orgs things nicely, better to include when things start to get complicated 
```

```python
# creating font dictionaires 
font1 = {'family': 'sans-serif', 'color': 'blue', 'size':20}
font2 = {'family': 'monospace', 'color': 'green', 'size':14}
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(X,y)
ax.set_title('Total Growth Over Time', fontdict=font1) # adding font dic
ax.set_ylabel('Total Growth', fontdict=font2) # adding font dic
ax.set_xlabel('Years Since Start', fontdict=font2) # adding font dic
fig.tight_layout() # keep everything fitting into container - orgs things nicely, better to include when things start to get complicated 
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(X,y)
ax.set_title('Total Growth Over Time', fontdict=font1, loc = 'left') # adding font dic, and title alignment
ax.set_ylabel('Total Growth', fontdict=font2) # adding font dic
ax.set_xlabel('Years Since Start', fontdict=font2, loc = 'right')  # adding font dic, and title alignment
fig.tight_layout() # keep everything fitting into container - orgs things nicely, better to include when things start to get complicated 
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(X,y,marker = '*', color = 'indigo')
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(X,y, marker ='*', color = 'indigo', linestyle = '--', linewidth = 2) # for colour can use hex codes instead
```
