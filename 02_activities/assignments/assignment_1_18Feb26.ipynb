```python
# import lib
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
np.random.seed(613)
```

```python
tips = sns.load_dataset("tips")
print(tips)
```

```python
sns.lineplot(data = tips,
             x = 'total_bill',
             y = 'tip')
```

```python
sns.set_style('whitegrid')

sns.lineplot(data = tips,
             x = 'total_bill',
             y = 'tip')
```

```python
sns.set_style('darkgrid')

sns.lineplot(data = tips,
             x = 'total_bill',
             y = 'tip')
```

```python
tipgraph = sns.lineplot(data = tips,
                        x = 'total_bill',
                        y = 'tip')
tipgraph.set(title = 'Tips vs. Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip Amount ($)')
```

```python
fig = plt.subplots(figsize=(10,3))

tipgraph = sns.lineplot(data = tips,
                        x = 'total_bill',
                        y = 'tip',
                        color = 'hotpink',
                        linestyle = '--',
                        linewidth = '3',
                        marker = 'o',
                        markerfacecolor = 'indigo')
tipgraph.set(title = 'Tips vs. Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip Amount ($)')
```

```python
# add multiple variables
fig = plt.subplots(figsize=(10,3))

tipgraph = sns.scatterplot(data = tips,
                        x = 'total_bill',
                        y = 'tip',
                        style = 'time',
                        hue = 'day',
                        palette = ['purple', 'hotpink', 'blue', 'green'])
tipgraph.set(title = 'Tips vs. Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip Amount ($)')
```

```python
# pairplot
sns.pairplot(data = tips, hue = 'day')
```

```python
#plotly
# import lib
import plotly.graph_objects as go
import nbformat
x1 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240,99, 220])
```

```python
graph = go.Figure() # graph obj
graph.add_trace(go.Bar(x=x1,y=y1))
graph.update_layout(title = "Pirate Scores",
                    xaxis_title = "Pirates",
                    yaxis_title = "Score")

graph.show()

#graph.write_html
```

```python
graph = go.Figure() # graph obj
graph.add_trace(go.Scatter(x=x1,y=y1, mode='markers', marker = dict(size = 15, color = 'hotpink', opacity = 1, line = dict(width = 5, color ='purple'))))

graph.update_layout(title = "Interactive Pirate Scores",
                    xaxis_title = "Pirates",
                    yaxis_title = "Score",
                    width = 500,
                    height = 500)

graph.show()
```

```python
#import wordcloud
#import lib
#ASSIGMNET 3 RELOOK
from wordcloud import WordCloud

df = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv", on_bad_lines= 'skip')

df
```

```python
# make string word cloud
text = " ".join(each for each in df.quote)

wordcloud = WordCloud(background_color= 'white',
                      colormap= 'inferno',
                      random_state= 613).generate(text)
fig, ax = plt.subplots(figsize = (7,3))
ax.imshow(wordcloud)
ax.axis("off")
```

```python
#import lib
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted

A = set(["apple", "banana", "watermelon"])
B = set(["pumpkin", "blueberry", "apple", "key lime"])

diagram = venn2([A, B],
                set_labels = ("Fruits", "Pies"),
                set_colors = ('blue', 'red'),
                alpha = 0.5)

diagram.get_label_by_id("10").set_text("\n".join(A - B)) # LEFT
diagram.get_label_by_id("11").set_text("\n".join(A & B)) # Middle
diagram.get_label_by_id("01").set_text("\n".join(B - A)) # Right 
```
